```python
import pandas as pd
import matplotlib as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house_price = pd.read_csv('train.csv')
```

```python
x_axis = house_price[['GrLivArea','BedroomAbvGr','FullBath']]
y_axis = house_price['SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_axis, y_axis, test_size=0.2, random_state=42)
```

```python
Predict_model = LinearRegression()
Predict_model.fit(X_train, y_train)
y_prediction = Predict_model.predict(x_axis)
residuals = y_axis - y_prediction
```

```python
Mean_squared_error = mean_squared_error(y_axis,y_prediction)
R_squared = r2_score(y_axis,y_prediction)
print(f'Validation Mean Squared Error: {Mean_squared_error: 0.2f}')
print(f'Validation R-squared: {R_squared: 0.2f}')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_axis, y=y_prediction)
plt.title('Actual vs. Predicted Values (Validation Set)')
plt.xlabel('Actual Sale Price')
plt.ylabel('Predicted Sale Price')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(residuals, kde=True)
plt.title('Distribution of Residuals (Validation Set)')
plt.xlabel('Residuals')
plt.show()
```

```python
# Predict house price 

new_data_point = [[1000, 4, 2]]  #square footage, BedroomAbvGr, FullBath
predicted_price = Predict_model .predict(new_data_point)
print(f'Predicted Price: ${predicted_price[0]}')
```
